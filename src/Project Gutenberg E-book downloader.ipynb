# Dutch books dump from Project Gutenberg
Execute all steps in order to get a file named 'allbooks-utf8.txt' containing all books from Project Gutenberg written in the Dutch language. 

Special thanks to: https://github.com/CaptainValor/gpt-2-colab

```
mkdir books
```

Download dutch books and unzip (takes a while)

```
!wget -w 0.5 -m -H "http://www.gutenberg.org/robot/harvest?filetypes[]=txt&langs[]=nl"
```

```
!find . -name "*[!-8].zip" | while read filename; do unzip -o -d "`basename -s .zip "$filename"`" "$filename"; done;
```

Collect .txt-files only 

```
!find . -name "*.txt" | while read filename; do cp $filename /content/books/; done;
```

Append each book with `<|endoftext|>` token

```
!find . -name "*.txt" | while read filename; do echo "<|endoftext|>" >> $filename; done;
```

Change charset to utf-8

```
!find . -name "*.txt" | while read filename; do iconv -f ascii -t utf8 $filename > $filename-utf8.txt ; done;
```

Combine all  files into single .txt-file

```
cat books/*utf8.txt >> allbooks-utf8.txt 
```
